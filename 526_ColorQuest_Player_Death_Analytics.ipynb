```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Upload your CSV file to Google Colab and update the file path
file_path = "/content/Alpha Build CSV.csv"

# Read the CSV file
df = pd.read_csv(file_path)

# Round the x-coordinate values to the nearest multiple of 5
df['x-co-ordinate'] = np.round(df['x-co-ordinate'] / 5) * 5

# Group the data by rounded x-coordinate and count the occurrences of each level
grouped_data = df.groupby('x-co-ordinate')['level'].value_counts().unstack(fill_value=0)
print(grouped_data)

# Define marker styles for each level
marker_styles = ['o', 's', 'D', 'v', '^', '<', '>', 'p', 'P', '*', 'h', 'H', '+', 'x', 'X', 'd']

# Plot each group of data with markers only for non-zero values
for i, level in enumerate(grouped_data.columns):
    # Filter non-zero values before plotting
    filtered_data = grouped_data[level][grouped_data[level] > 0]
    plt.plot(filtered_data.index, filtered_data.values, marker=marker_styles[i % len(marker_styles)], label=f'Level {level}')

# Add labels and title
plt.xlabel('x-coordinate (rounded to nearest multiple of 5)')
plt.ylabel('Number of Players dead')
plt.title('Number of Players Dead at x-coordinate (nearest multiple of 5)')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
